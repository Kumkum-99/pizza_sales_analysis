```python
import pandas as pd
pizzas = pd.read_csv("pizzas.csv", encoding="cp1252")
order_details = pd.read_csv("order_details.csv", encoding="cp1252")
orders = pd.read_csv("orders.csv", encoding="cp1252")
pizza_types = pd.read_csv("pizza_types.csv", encoding="cp1252")
```

```python
df=order_details.merge(pizzas,on='pizza_id',how='left')
df=df.merge(orders,on='order_id',how='left')
df=df.merge(pizza_types,on='pizza_type_id',how='left')
df.head()
```
